### Hadoop
Hadoop was set up locally in our environments through either docker containers or WSL.<br> Here are some of the commands used.

Places file into hadoop directory so MapReduce can be run

```python
hadoop fs -put /home/neha/testhadoop.csv /user/hadoop/input
```

Takes a file or file endings in a folder, as an input and runs it through Hadoop with both mapper and reducer files specified.

```python
# Needs environment variables to be set and paths will be different
$HADOOP_HOME/bin/hadoop jar $HADOOP_HOME/share/hadoop/tools/lib/hadoop-streaming-3.3.6.jar   -input /user/hadoop/input/*.csv   -output /user/hadoop/output/test1   -mapper "python3 /home/neha/mapper.py"   -reducer "python3 /home/neha/reducer.py"
```

Displays the output of the MapReduce

```python
hadoop fs -cat /user/hadoop/output/test1/part-00000
```
